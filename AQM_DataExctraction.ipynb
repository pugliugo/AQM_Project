```python
df_comm = ek.get_data(
    instruments = ['CLc1', 'LCOc1', 'GCc1', 'SIc1', 'PLc1', 'NGc1', 'Wc1', 'Sc1', 'HOc1', 'Cc1'],
    fields = ['TR.CLOSEPRICE'],
    parameters = {
        'Curn': 'Native',
        'SDate': '2012-03-01',
        'EDate': '2022-03-01',
        'Frq': 'M'
    }
)

display(df_comm)
```

```python
# Using the function get_data with the instruments would only output values in this order.
# It is first not handy to work with, and it would also outputs many missing values, thus making it difficult
# To resample the dataset.
```

```python
df_comm = ek.get_timeseries(['CLc1', 'LCOc1', 'GCc1', 'SIc1', 'PLc1', 'NGc1', 'Wc1', 'Sc1', 'HOc1', 'Cc1'], start_date='2012-03-01',end_date='2022-03-01',interval='monthly')

display(df_comm)
```

```python
df_comm.to_excel('Commodities.xlsx')
```
